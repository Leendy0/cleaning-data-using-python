```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# read the data into a pandas dataframe
df = pd.read_excel("Dots Potential Customer Survey Data.xlsx")

# create a feature matrix and target vector
X = df.drop(["Unique ID", "Annual Family Income ($)"], axis=1)
y = df["Country"]

# scale the numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X[["Age", "Time spent watching videos/TV"]])

# concatenate the scaled features with the categorical features
gender_dummies = pd.get_dummies(X["Gender"], prefix="Gender")

X_processed = pd.concat(
    [gender_dummies, pd.DataFrame(X_scaled, columns=["Age", "Sleeping hours"])],
    axis=1,
)

print(X_processed)
```
